```
import pandas as pd
import numpy as np
import pdb
from matplotlib import pyplot as plt
```

```
noOfActions = 10
meanValue = 0.0
varianceValue = 1.0

#epsilon-greedy logic
# others to be written - Optimistic Initial Value, Upper-Confidence_Bound, Gradient Method, non-stationary problem and associative search
epsilon = 0.1

numberOfRuns = 1000
numberOfTimeSteps = 1000

reward = {}
action = {}
trueValues = {}
isOptimalAction = {}
for runNumber in range(0,numberOfRuns):
    trueValues[runNumber] = pd.Series(np.random.normal(meanValue, varianceValue, noOfActions), index = range(0,noOfActions))
    bestAction = trueValues[runNumber][trueValues[runNumber] == trueValues[runNumber].max()].index[0] 
    action[runNumber] = {}
    reward[runNumber] = {}
    isOptimalAction[runNumber] = {}
    estimatedValues = pd.Series(0.0, index = range(0,noOfActions))
    numberofTypeActions = pd.Series(0.0, index = range(0,noOfActions))
    isEpsilon = {}
    for timeStep in range(0,numberOfTimeSteps):
        if np.random.random() < (1.0 - epsilon):
            xx = estimatedValues[estimatedValues == estimatedValues.max()]
            action[runNumber][timeStep] = xx.index[np.random.random_integers(0, len(xx) - 1.0)] if len(xx) == 1 else xx.index[0]
            isEpsilon[timeStep] = False
        else:
            isEpsilon[timeStep] = True
            action[runNumber][timeStep] = trueValues[runNumber].index[np.random.random_integers(0, len(trueValues[runNumber]) - 1.0)]
        isOptimalAction[runNumber][timeStep] = (action[runNumber][timeStep] == bestAction)
        reward[runNumber][timeStep] = np.random.normal(trueValues[runNumber][action[runNumber][timeStep]], varianceValue)
        estimatedValues[action[runNumber][timeStep]] = (estimatedValues[action[runNumber][timeStep]]*numberofTypeActions[action[runNumber][timeStep]] + reward[runNumber][timeStep])*1.0/(numberofTypeActions[action[runNumber][timeStep]] + 1.0)
        numberofTypeActions[action[runNumber][timeStep]] = numberofTypeActions[action[runNumber][timeStep]] + 1.0
    test = pd.Series(isEpsilon)
    print (runNumber, 'finished', len(test[test])*1.0/numberOfTimeSteps, 'Percentage of Epsilon Events')
averageReward = pd.DataFrame(reward).mean(axis = 1)
percOptAction = pd.DataFrame(isOptimalAction).sum(axis = 1)*1.0/numberOfRuns
```

```
averageReward.plot()
plt.show()
```
