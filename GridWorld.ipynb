```
import pandas as pd
import numpy as np
import pdb
from matplotlib import pyplot as plt
from copy import deepcopy
```

```
worldSize = 4
reward = -1
actionProb = 0.25

def getStates(row, col):
  if (row == 0 and col == 0) or (row == (worldSize - 1) and (col == (worldSize - 1))):
    allStates = pd.DataFrame([[0, 0]], index = [0], columns = ['row', 'col'])
    rewards = pd.Series([0.0], index = [0])
    allProbs = pd.Series([1.0], index = [0])
    return allStates, rewards, allProbs
  prevCell = [row, col - 1]
  nextCell = [row, col + 1]
  upperCell = [row - 1, col]
  lowerCell = [row + 1, col]
  if row == 0:
    upperCell = [row, col]
  if row == 3:
    lowerCell = [row, col]
  if col == 0:
    prevCell = [row, col]
  if col == 3:
    nextCell = [row, col]
  allStates = pd.DataFrame([prevCell, nextCell, upperCell, lowerCell], index = range(0, 4), columns = ['row', 'col'])
  rewards = pd.Series(-1.0, index = range(0, len(allStates)))
  allProbs = pd.Series(1.0, index = range(0, len(allStates)))
  return allStates, rewards, allProbs



valueDf = pd.DataFrame(0.0, index = range(0, worldSize), columns = range(0, worldSize))
while True:
  currValueDf = deepcopy(valueDf)
  for row in range(0, worldSize):
    for col in range(0, worldSize):
      allStates, allRewards, allProbs = getStates(row, col)
      valueDf[col][row] = (pd.Series([allProbs[i]*(allRewards[i] + currValueDf[allStates['col'][i]][allStates['row'][i]]) for i in allStates.index])*actionProb).sum()
  delta = (valueDf - currValueDf).abs().sum().sum()
  if delta < 1e-8:
    break
print valueDf
      
```

```
#q(11, down)
def calculateQfromV(row, col, finalRow, finalCol):
  return reward + valueDf[finalRow][finalCol]
#print calculateQfromV(2, 3, 3, 3)
print calculateQfromV(1, 3, 2, 3)
print valueDf
```
